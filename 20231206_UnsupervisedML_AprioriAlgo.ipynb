```python
#Apriori preparation
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
#Support code
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```python
#Support with column names
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

Conclusion: 
80% des clients ont acheté un pantalon.
80% des clients ont acheté une écharpe

```python
#Confidence code:
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
```

Interpretation:
Ligne 2 : 100% dans paniers qui contiennent une jupe contiennent un pantalon.
A l'inverse, 75% des paniers qui contiennent un pantalon contiennent  une jupe

Interprétation antecedent support: 80% des paniers contiennent des écharpes, pantalons, et 60% contiennent des jupes
Interprétation consequent support: même interprétation que pour la colonne précedente : 80% des paniers contiennent pantalons et écharpes, 60% contiennent jupe.

Support : c'est la combinaison des deux : 
60% des paniers contiennent à la fois écharpe et pantalon
60% des paniers contiennent à la fois jupe et pantalon

Confidence: 
ligne 2 : 100%  des  client qui ont acheté une  jupe  ont un pantalon
75% des clients qui ont acheté un pantalon ont acheté une jupe.

Lift : il y a plus de chance que les clients achètent des pantalons et  jupes ensemble  plutôt que séparemment.
En revanche,  il y a plus de chance que les clients achètent un pantalon et écharpe séparemment plutot que ensemble.
