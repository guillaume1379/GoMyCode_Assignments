1.2) Write a function to Extract article title

```python
def ExtractTitle():
    link = str(input('input your url'))
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    title = soup.select('title')[0].getText()
    print(title)
```
