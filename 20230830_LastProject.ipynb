```python
import tkinter as tk
import random
```

### Game Initialization

```python
#Initializing game window
window = tk.Tk()
#window.geometry("300x250")
window.config(bg="grey")
window.title('Rock Paper Scissors Game')


#initialisation des valeurs de scores
CompteurVictoires = 0
CompteurDefaites = 0
CompteurNuls = 0

#On crée le compteur si on fait plusieurs parties

def Gamer_choiceRock():
    LabelChoiceGamer['text'] = 'Rock'
    
def Gamer_choicePaper():
    LabelChoiceGamer['text'] = 'Paper'
    
def Gamer_choiceScissors():
    LabelChoiceGamer['text'] = 'Scissors'

def Reset():
    global CompteurVictoires, CompteurDefaites, CompteurNuls
    CompteurVictoires = 0
    CompteurDefaites = 0
    CompteurNuls = 0
    ScoreJoueur['text'] = CompteurVictoires
    ScoreCPU['text'] = CompteurDefaites 
    ScoreDraw['text'] = CompteurNuls
    
def LaunchTheGame():
    global CompteurVictoires, CompteurDefaites, CompteurNuls
    #CPU va choisir aléatoirement son choix
    LabelChoiceComputer['text'] = random.choice(['Rock','Paper','Scissors']) 
    Choix = LabelChoiceComputer['text']
    
    #Lancement du jeu
    choiceGamer = LabelChoiceGamer['text']
    choiceComputer = Choix  
    if (choiceComputer == 'Rock') and (choiceGamer == 'Paper'):
        OutputGame = 'Win'
    elif (choiceComputer == 'Paper') and (choiceGamer == 'Scissors'):
        OutputGame == 'Win'
    elif (choiceComputer == 'Scissors') and (choiceGamer == 'Rock'):
        OutputGame = 'Win'
    elif (choiceComputer == choiceGamer):
        OutputGame = 'Draw'
    else:
        OutputGame = 'Lose'
    
    #Affichage résultat du jeu
    LabelGameResult['text'] = OutputGame
    
    #Mettre en place le compteur
    if OutputGame == 'Win':
        CompteurVictoires = CompteurVictoires + 1
    elif OutputGame == 'Draw':
        CompteurNuls = CompteurNuls + 1
    else:
        CompteurDefaites = CompteurDefaites + 1
    
    #affichage compteur
    ScoreJoueur['text'] = CompteurVictoires
    ScoreCPU['text'] = CompteurDefaites 
    ScoreDraw['text'] = CompteurNuls 
     

def Close():
    window.destroy()
    
#Graphisme de ma page        
        
Label0 = tk.Label(master = window , text= 'Welcome to Rock Paper Scissors Game!')
Label0.grid(row = 0,column= 0)

Label1 = tk.Label(master = window , text='Select your item for the game')
Label1.grid(row = 2 , column = 0)

Button1 = tk.Button(master = window , text = 'Rock', command = Gamer_choiceRock)
Button1.grid(row=3,column=0)

Button2 = tk.Button(master = window , text = 'Paper', command = Gamer_choicePaper)
Button2.grid(row=3,column=1)

Button3 = tk.Button(master = window , text = 'Scissors', command = Gamer_choiceScissors)
Button3.grid(row=3,column=2)

Label2 = tk.Label(master = window , text = 'The choice you have selected for the game is:')
Label2.grid(row = 4, column = 0)

LabelChoiceGamer = tk.Label(master = window , fg = 'dark blue')
LabelChoiceGamer.grid(row=4,column=1)

Label3 = tk.Label(master = window, text = 'Click here to launch the game!')
Label3.grid(row=6,column = 0)

#ButtonLaunchGame = tk.Button(master = window, text='LAUNCH!', command=lambda:[LabelChoiceComputer(),LaunchGameTest()])
ButtonLaunchGame = tk.Button(master = window, fg = 'red', text='LAUNCH!', command=LaunchTheGame)
ButtonLaunchGame.grid(row=6 , column=1)

#Il faut mettre le choix de l'ordinateur
LabelChoiceComputerText = tk.Label(master=window, text = 'Computer chose :')
LabelChoiceComputerText.grid(row=7, column=0)

LabelChoiceComputer = tk.Label(master = window)
LabelChoiceComputer.grid(row=7,column = 1)

#Affichage du résultat du jeu
LabelGameResultText = tk.Label(master=window, text = 'Output of the game :')
LabelGameResultText.grid(row=9, column=0)

LabelGameResult = tk.Label(master = window)
LabelGameResult.grid(row = 9, column=1)

#il faut ensuite une fenêtre de score
LabelScoreJoueurText = tk.Label(master=window, text = 'User Score')
LabelScoreJoueurText.grid(row = 10 , column=0)

ScoreJoueur = tk.Label(master = window)
ScoreJoueur.grid(row=10,column=1)

LabelScoreCPUText = tk.Label(master=window, text = 'CPU Score')
LabelScoreCPUText.grid(row = 11 , column=0)

ScoreCPU = tk.Label(master = window)
ScoreCPU.grid(row=11,column=1)

LabelScoreDrawText = tk.Label(master=window, text = 'Draw Score')
LabelScoreDrawText.grid(row = 12 , column=0)

ScoreDraw = tk.Label(master = window)
ScoreDraw.grid(row=12,column=1)

# Button for resetting the results
reset_button = tk.Button(master=window, text="Reset Results", command=Reset)
reset_button.grid(row=13,column=0)

# Button for closing
exit_button = tk.Button(master=window, text="Exit", command=Close)
exit_button.grid(row=14,column=3)


window.mainloop()
```
